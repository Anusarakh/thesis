```python
import librosa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
import csv
```

```python
y, sr = librosa.load(librosa.ex('libri1'))
mfcc_test = librosa.feature.mfcc(y=y, sr=44100, S=None, n_mfcc=20)
```

```python
df_ = pd.DataFrame(mfcc_test)
df_
```

```python
mfcc_mean = mfcc_test.mean()
mfcc_mean
```

```python
zcr = librosa.feature.zero_crossing_rate(y)
zcr
```

```python
df_zero = pd.DataFrame(zcr)
df_zero
```

```python
zcr_mean = zcr.mean()
zcr_mean
```

```python
import matplotlib.pyplot as plt
import scipy
```

```python
lpc_test = librosa.lpc(y, order=1, axis=-1)
lpc_test
```

```python
df_lpc_test = pd.DataFrame(lpc_test)
df_lpc_test
```

```python
lpc_test_mean = lpc_test.mean()
lpc_test_mean
```

```python
spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
spec_cent
```

```python
spec_cent_mean = spec_cent.mean()
spec_cent_mean
```

```python
df_spec_cent = pd.DataFrame(spec_cent)
df_spec_cent
```
