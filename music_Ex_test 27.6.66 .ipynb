#MFCC

```python
mfcc = librosa.feature.mfcc(y=y, sr=44100 ,n_mfcc= 20, n_fft = 512,  hop_length=256 )
mfcc
```

```python
df_mfcc = pd.DataFrame(mfcc)
df_mfcc
```

```python
df_mfcc.mean(axis = 1) # MFCC
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows=2, sharex=True)
img = librosa.display.specshow(librosa.power_to_db(mfcc, ref=np.max),
                               x_axis='time', y_axis='mel', fmax=8000,
                               ax=ax[0])
fig.colorbar(img, ax=[ax[0]])
ax[0].set(title='Mel spectrogram')
ax[0].label_outer()
img = librosa.display.specshow(mfcc, x_axis='time', y_axis='mel', fmax=8000, ax=ax[1])
fig.colorbar(img, ax=[ax[1]])
ax[1].set(title='MFCC')
```

# (Spectral Centroid)

```python
SC = librosa.feature.spectral_centroid(y=y, sr=44100, S=None, n_fft=512, hop_length=256)
df_SC = pd.DataFrame(SC)
df_SC
```

```python
import matplotlib.pyplot as plt
times = librosa.times_like(SC)
fig, ax = plt.subplots()
librosa.display.specshow(librosa.amplitude_to_db(SC, ref=np.max),
                         y_axis='log', x_axis='time', ax=ax)
ax.plot(times, SC.T, label='Spectral centroid', color='w')
ax.legend(loc='upper right')
ax.set(title='log Power spectrogram')
```
